```python
import pandas as pd
```

```python
data=pd.read_csv('customer_shopping_behavior.csv')
data.head()
```

```python
(data.isnull().sum()/len(data))*100
```

```python
data[data['Review Rating'].isnull()]
```

```python
data.groupby('Category')['Review Rating'].median()
```

```python
#removing null values
data['Review Rating']=data.groupby('Category')['Review Rating'].transform(lambda x:x.fillna(x.median()))
```

```python
#changing the cloumn in snake_case format
data.columns=data.columns.str.lower()
data.columns=data.columns.str.replace(' ','_')
data=data.rename(columns={'purchase_amount_(usd)':'purchase_amount'})
```

```python
#creating new column age_group
labels=['Young Adult','Adult','Middle-Aged','Senior']
data['age_group']=pd.qcut(data['age'],q=4,labels=labels)
data[['age','age_group']].head(10)
```

```python
#creating column purchase frequency days
frequency_mapping={
    'Fortnightly':14,
    'Bi-Weekly':14,
    'Weekly':7,
    'Monthly':30,
    'Quarterly':90,
    'Every 3 Months':90,
    'Annually':365,
}
data['purchase_frequency_days']=data['frequency_of_purchases'].map(frequency_mapping)
```

```python
data[['purchase_frequency_days','frequency_of_purchases']].head(20)
```

```python
data=data.drop(columns='purchase_frequency_days')
```

```python
data[['promo_code_used','discount_applied']]
```

```python
#checking both column have same data
(data['discount_applied']==data['promo_code_used']).all()#contains same so we need to drop any one column
```

```python
data.drop(columns='promo_code_used', inplace=True)
```
